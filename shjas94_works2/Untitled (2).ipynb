```python
import json
```

```python
with open('./train.json') as json_file1:
    json_data1 = json.load(json_file1)
with open('./test.json') as json_file2:
    json_data2 = json.load(json_file2)
```

```python
start = 2617
a = json_data1['images']
b = json_data2['images']
for i in b:
    i['id'] = start
    start+=1
img = a+b
print(img)

```

```python
import pandas as pd
import numpy as np
PredictionString = pd.read_csv('./submission_jkw_ensemble_wbf_2.csv')['PredictionString']
annotation = []
ids = 21116
image_id = 2617
for bboxs in PredictionString:
    bboxs = bboxs.strip().split(' ')
    bboxs = [float(i) for i in bboxs]
    bboxs = [bboxs[i:i + 6] for i in range(0, len(bboxs), 6)]
    for bbox in bboxs:
        prob = bbox[1]
        if prob<0.95:
            print(prob)
            continue
        temp = dict()
        temp['id'] = ids
        temp['image_id'] = image_id
        temp['category_id'] = int(bbox[0])
        temp['segmentation'] = [[1,2,3,4,5]]
        temp['area'] = 1
        temp['bbox'] = [i for i in list(np.round(bbox[2:], 1))]
        if len(bbox)!=6:
            print(bbox)
        temp['iscrowd'] = 0
        annotation.append(temp)
        ids+=1
    image_id+=1

    
```

```python
json_data1['images'] = img
json_data1['annotations'] = json_data1['annotations']+annotation
```

```python
with open('./train+pseudo.json', 'w') as outfile:
    json.dump(json_data1, outfile,indent=4)
```
